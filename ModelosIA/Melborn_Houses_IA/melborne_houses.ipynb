# Projeto de Machine Learning para predição de preços de casas em Melborne

```python
# Importando base de dados e dependências

import pandas as pd

base = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Melborn_Houses/melb_data.csv')

display(base.info())
```

```python
# iniciando tratamento dos dados

# Colunas com muitos valores nulos
display(base['BuildingArea'].isnull().value_counts())
display(base['YearBuilt'].isnull().value_counts())
display(base['CouncilArea'].isnull().value_counts())

#display(base.Car)
base.info()
```

```python

# Desmembrando a coluna Type em numeros inteiros para possuir mais dados de treinamento
dummies_type = pd.get_dummies(base.Type, dtype=int)
base = pd.concat([base, dummies_type], axis=1).drop('Type', axis=1)


display(base)
display(base.info())
```

```python
# Desmembrando a coluna Method em numeros inteiros para possuir mais dados de treinamento
dummies_method = pd.get_dummies(base.Method, dtype=int)
base = pd.concat([base, dummies_method], axis=1).drop('Method', axis=1)


display(base)
display(base.info())
```

```python

# Desmembrando a coluna Regionname em numeros inteiros para possuir mais dados de treinamento
dummies_regionname = pd.get_dummies(base.Regionname, dtype=int)
base = pd.concat([base, dummies_regionname], axis=1).drop('Regionname', axis=1)


display(base)
display(base.info())
```

```python

remover = ['Suburb', 'Date', 'Address', 'SellerG', 'BuildingArea', 'YearBuilt', 'CouncilArea']
base = base.drop(remover, axis=1)

display(base)
base.info()
```

```python
# Tratando a coluna 'Car' para substituir os valores nulos
print("\nColuna Car antes:\n")
display(base.Car.info())
base.loc[base.Car.isnull() == True] = base.Car.median()
print("\nColuna Car depois:\n")
display(base.Car.info())
display(base.Car)
```

```python
# Iniciando treinamento do modelo

x = base.drop('Price', axis=1)
y = base.Price

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.33, random_state=42)

display(x_treino.info())

reg  = LinearRegression()

reg.fit(x_treino, y_treino)
```

```python
# Verificando a adaptabilidade dos dados

print(f'Score do modelo: {reg.score(x_treino, y_treino)}')

pred = reg.predict(x_teste)

from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

erro_absoluto = mean_absolute_error(y_teste, pred)

erro_quadrado = mean_squared_error(y_teste, pred)

raiz_erro = root_mean_squared_error(y_teste, pred)

r2 = r2_score(y_teste, pred)

print(f"Erro absoluto do modelo: {erro_absoluto}")

print(f"Quadrado erro absoluto do modelo: {erro_quadrado}")

print(f"Raiz quadrada do erro absoluto do modelo: {raiz_erro}")

print(f"R2 Score do modelo: {r2}")
```

```python
# Utilizando agora arvore de decisão regressora

from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor()

tree.fit(x_treino, y_treino)

pred_tree = tree.predict(x_teste)

print(f'Score do modelo: {reg.score(x_treino, y_treino)}')

erro_absoluto_tree = mean_absolute_error(y_teste, pred_tree)

erro_quadrado_tree = mean_squared_error(y_teste, pred_tree)

raiz_erro_tree = root_mean_squared_error(y_teste, pred_tree)

r2_tree = r2_score(y_teste, pred_tree)

print(f"Erro absoluto do modelo: {erro_absoluto_tree}")

print(f"Quadrado erro absoluto do modelo: {erro_quadrado_tree}")

print(f"Raiz quadrada do erro absoluto do modelo: {raiz_erro_tree}")

print(f"R2 Score do modelo: {r2_tree}")
```
