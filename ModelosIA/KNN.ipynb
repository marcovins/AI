# Implementando algoritmo de machine learning em dataset íris

```python
# Importando base de dados

from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
base = pd.DataFrame(iris.data, columns=iris.feature_names)
base['target'] = iris.target

display(base)
```

```python
# Imporando modelo do KNN

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

# Separando apenas colunas de petalas para melhor visualização

base1 = base.iloc[:, 2:5].loc[base['target'] != 0]
display(base1)
```

```python
# Separando X e Y

x = base1.drop('target', axis=1)
y = base1['target']

# Separando dados entre treino e teste

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.33, random_state=42)

# Treinando modelo

knn.fit(x_treino, y_treino)

# Fazendo previsões

previsoes = knn.predict(x_teste)

# Avaliando previsões com a matriz de precisão

from sklearn.metrics import confusion_matrix

print(f"\nMatriz de confusão: \n{confusion_matrix(y_teste, previsoes)}")
```
