# Regressão no Scikit-Learn e deploy do modelo

```python
# Importando base de dados

import pandas as pd


base_treino = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Casas/casas_treino.csv')



base_treino = base_treino.drop(remover, axis=1)


# Definindo X e Y para treino dos modelos

x_treino = base_treino.drop('MedHouseVal', axis=1)
y_treino = base_treino['MedHouseVal']

base_treino.info()
```

```python
# Iniciando treinamento do modelo

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_treino, y_treino)
```

```python
# Testando modelo utilizando score

base_teste = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Casas/casas_teste.csv')

base_teste = base_teste.drop(remover, axis=1)

x_teste = base_teste.drop('MedHouseVal', axis=1)
y_teste = base_teste['MedHouseVal']

reg.score(x_teste, y_teste)
```

```python
# Exibindo coeficiente e intercept
print("Coeficientes: ")
print(reg.coef_)
print("Intercept: ")
print(reg.intercept_)
```

# Implementando persistência do modelo

```python
# Importando joblib

from joblib import dump, load

dump(reg, '/content/drive/MyDrive/Colab Notebooks/Casas/reg.joblib')

# Importando arquivo para verificar

reg_load = load('/content/drive/MyDrive/Colab Notebooks/Casas/reg.joblib')

print("Coeficientes: ")
print(reg_load.coef_)

print("Intercept: ")
print(reg_load.intercept_)
```

```python
# Salvando previsões em arquivo de excel

previsao = reg_load.predict(x_teste)

base_teste['Previsto'] = previsao

base_teste.to_excel('/content/drive/MyDrive/Colab Notebooks/Casas/previsoes.xlsx')

display(base_treino)
```
