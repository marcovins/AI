```python
# Passo 1 - Percorrer todos os arquivos da pasta base de dados

import os
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots


lista_arquivos = os.listdir("/content/drive/MyDrive/Colab Notebooks/Vendas")
display(lista_arquivos)
```

```python
# Passo 2 - Importar as bases de dados de vendas

tabela_completa = pd.DataFrame()

for arquivo in lista_arquivos:

  if "Vendas" in arquivo:

    tabela = pd.read_csv(f"/content/drive/MyDrive/Colab Notebooks/Vendas/{arquivo}")

    tabela_completa = pd.concat([tabela_completa, tabela])
```

```python
# Passo 3 - Tratar/Compilar as bases de dados

display(tabela_completa)
```

```python
# Passo 4 - Calcular o produto mais vendido (em quantidade)

tabela_produtos = tabela_completa.groupby('Produto').sum()
tabela_vendas = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
display(tabela_vendas)
```

```python
# Passo 5 - Calcular o produto que mais faturou (em faturamento)

tabela_completa['Faturamento'] = tabela_completa['Quantidade Vendida'] * tabela_completa['Preco Unitario']
tabela_faturamento = tabela_completa[['Produto', 'Faturamento']].groupby('Produto').sum().sort_values(by='Faturamento', ascending=False)
display(tabela_faturamento)
```

```python
# Passo 6 - Calcular a loja/cidade que mais (vendeu em faturamento)

tabela_cidades = tabela_completa.groupby('Loja')['Faturamento'].sum()
tabela_cidades = tabela_cidades.sort_values(ascending=False)

display(tabela_cidades)
```

```python
# Dashboard produto mais vendido

fig_venda = px.bar(tabela_vendas, x=tabela_vendas.index, y='Quantidade Vendida', title='Produtos mais vendidos')
fig_venda.show()
```

```python
# Dashboard produto que mais faturou

fig_fatura = px.bar(tabela_faturamento, x=tabela_faturamento.index, y=tabela_faturamento['Faturamento'], title='Produtos que mais faturaram')
fig_fatura.show()
```

```python
# Dashboard cidade que mais faturou

fig_cidade = px.bar(tabela_cidades, x=tabela_cidades.index, y=tabela_cidades, title='Cidades que mais faturaram')
fig_cidade.show()
```
